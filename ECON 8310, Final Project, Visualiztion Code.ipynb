```python
import pandas as pd
health = pd.read_csv('https://raw.githubusercontent.com/anvesh232/grp_econ_busfor/refs/heads/main/health_dataset.csv')
survey= pd.read_csv('https://raw.githubusercontent.com/anvesh232/grp_econ_busfor/refs/heads/main/survey.csv')
```

```python
import pandas as pd
# Renaming columns
survey.rename(columns={
    'Survey ResponseID': 'Survey.ResponseID','Q-demos-age': 'Q.demos.age','Q-demos-hispanic': 'Q.demos.hispanic','Q-demos-race': 'Q.demos.race','Q-demos-education': 'Q.demos.education','Q-demos-income': 'Q.demos.income','Q-demos-gender': 'Q.demos.gender','Q-sexual-orientation': 'Q.sexual.orientation','Q-demos-state': 'Q.demos.state','Q-amazon-use-howmany': 'Q.amazon.use.howmany','Q-amazon-use-hh-size': 'Q.amazon.use.hh.size','Q-amazon-use-how-oft': 'Q.amazon.use.how.oft','Q-substance-use-cigarettes': 'Q.substance.use.cigarettes','Q-substance-use-marijuana': 'Q.substance.use.marijuana','Q-substance-use-alcohol': 'Q.substance.use.alcohol','Q-personal-diabetes': 'Q.personal.diabetes','Q-personal-wheelchair': 'Q.personal.wheelchair','Q-life-changes': 'Q.life.changes','Q-sell-YOUR-data': 'Q.sell.YOUR.data','Q-sell-consumer-data': 'Q.sell.consumer.data','Q-small-biz-use': 'Q.small.biz.use','Q-census-use': 'Q.census.use','Q-research-society': 'Q.research.society'
}, inplace=True)
```

```python
import pandas as pd
# Merging data using Survey.ResponseID
combined_data = health.merge(survey, how='left', on='Survey.ResponseID')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Plot the bar chart for the age range distribution of health product buyers
combined_data['Q.demos.age'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')

# Set the title of the plot
plt.title("Age Range Distribution of Health Product Buyers")

# Set the label for the x-axis
plt.xlabel("Age Range")

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Set the label for the y-axis
plt.ylabel("Count")

# Show the plot
plt.show()

```

```python
# Calculate the count of each age range and reset the index
age_distribution = combined_data['Q.demos.age'].value_counts().reset_index()

# Rename columns for better readability
age_distribution.columns = ['Q.demos.age', 'count']

# Calculate percentage for each age range
age_distribution['percentage'] = (age_distribution['count'] / age_distribution['count'].sum()) * 100

# Plot the bar chart for age range percentage
plt.bar(age_distribution['Q.demos.age'], age_distribution['percentage'], color='lightgreen', edgecolor='black')

# Set the title of the plot
plt.title("Age Range Percentage of Health Product Buyers")

# Set the label for the x-axis
plt.xlabel("Age Range")

# Set the label for the y-axis
plt.ylabel("Percentage")

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Adjust the layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()

```

```python
import seaborn as sns

# Create a figure with specified size
plt.figure(figsize=(10, 6))

# Plot the countplot with age and gender distribution
sns.countplot(data=combined_data, x='Q.demos.age', hue='Q.demos.gender', dodge=True)

# Set the title of the plot
plt.title("Age and Gender Distribution")

# Set the label for the x-axis
plt.xlabel("Age Range")

# Set the label for the y-axis
plt.ylabel("Count")

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Adjust the layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Group the data by age and gender, then count the occurrences
gender_age_distribution = combined_data.groupby(['Q.demos.age', 'Q.demos.gender']).size().reset_index(name='count')

# Calculate the percentage for each gender within each age range
gender_age_distribution['percentage'] = gender_age_distribution.groupby('Q.demos.age')['count'].transform(lambda x: x / x.sum() * 100)

# Create a figure with specified size
plt.figure(figsize=(10, 6))

# Plot the bar chart with seaborn
sns.barplot(data=gender_age_distribution, x='Q.demos.age', y='percentage', hue='Q.demos.gender', dodge=True)

# Set the title of the plot
plt.title("Relative Gender Percentages Within Each Age Range")

# Set the label for the x-axis
plt.xlabel("Age Range")

# Set the label for the y-axis
plt.ylabel("Percentage")

# Add a legend with the title "Gender"
plt.legend(title="Gender")

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Adjust the layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the count of each income bracket and reset the index
income_counts = combined_data['Q.demos.income'].value_counts().reset_index()

# Rename columns for better readability
income_counts.columns = ['Q.demos.income', 'count']

# Sort the income counts in descending order
income_counts = income_counts.sort_values(by='count', ascending=False)

# Create a figure with specified size
plt.figure(figsize=(10, 6))

# Plot the bar chart with seaborn
sns.barplot(data=income_counts, x='Q.demos.income', y='count')

# Set the title and labels for the chart
plt.title("Income Levels of Buyers")
plt.xlabel("Income Bracket")
plt.ylabel("Count")

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Adjust the layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert the 'Q.amazon.use.how.oft' column to string type
combined_data['Q.amazon.use.how.oft'] = combined_data['Q.amazon.use.how.oft'].astype(str)

# Get the ordered data based on the frequency of usage
ordered_data = combined_data['Q.amazon.use.how.oft'].value_counts().index

# Create a figure with specified size
plt.figure(figsize=(10, 6))

# Plot the countplot with seaborn
sns.countplot(data=combined_data, x='Q.amazon.use.how.oft', order=ordered_data, color='orange')

# Set the title of the plot
plt.title("Amazon Usage Frequency of Buyers")

# Set the label for the x-axis
plt.xlabel("Frequency")

# Set the label for the y-axis
plt.ylabel("Count")

# Adjust the layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
